## 3. Explorar y preparar los datos

Revisa la estructura del DataFrame y convierte las variables categóricas en variables dummy.

```python
df = pd.get_dummies(df, columns=['InternalName'], drop_first=True)
```

## 4. Definir variables dependiente e independientes

Selecciona las variables que utilizarás en el modelo.

```python
# Variable dependiente
y = df['NumFirstConnect']

# Variables independientes: 'Age' y las dummies de 'InternalName' y 'Modem'
modem_cols = [col for col in df.columns if col.startswith('Modem_')]
internalname_cols = [col for col in df.columns if col.startswith('InternalName_')]
X = df[['Age'] + modem_cols + internalname_cols]
```

## 5. Agregar constante y ajustar el modelo

```python
# Agregar constante al modelo
X = sm.add_constant(X)

# Ajustar el modelo de regresión lineal múltiple
model = sm.OLS(y, X).fit()
```
